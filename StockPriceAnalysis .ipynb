# Overlap Studies Functions

Reference: https://mrjbq7.github.io/ta-lib/func_groups/overlap_studies.html

```python
overlap_df = pd.DataFrame({"Date": df["Date"], "Closing Prices": df["Adj Close"]})
overlap_df.head(5)
```

```python
upperband, middleband, lowerband = BBANDS(closing_prices, timeperiod=5, nbdevup=2, nbdevdn=2, matype=0)
overlap_df["BBands.upper"] = upperband
overlap_df["BBands.middle"] = middleband
overlap_df["BBands.lower"] = lowerband
overlap_df.head(10)
```

```python
dema = DEMA(closing_prices, timeperiod=5)
overlap_df["DEMA"] = dema
overlap_df.head(10)
```

```python
ema = EMA(closing_prices, timeperiod=5)
overlap_df["EMA"] = ema
overlap_df.head(10)
```

```python
ht_trendline = HT_TRENDLINE(closing_prices)
overlap_df["HT_TRENDLINE"] = ht_trendline
overlap_df.head(10)
```

```python
kama = KAMA(closing_prices, timeperiod=5)
overlap_df["KAMA"] = kama
overlap_df.head(10)
```

```python
ma = MA(closing_prices, timeperiod=5, matype=0)
overlap_df["MA"] = ma
overlap_df.head(10)
```

```python
midpoint = MIDPOINT(closing_prices, timeperiod=5)
overlap_df["MIDPOINT"] = midpoint
overlap_df.head(10)
```

```python
midprice = MIDPRICE(df["High"], df["Low"], timeperiod=5)
overlap_df["MIDPRICE"] = midprice
overlap_df.head(10)
```

```python
sma = SMA(closing_prices, timeperiod=5)
overlap_df["SMA"] = sma
overlap_df.head(10)
```

```python
t3 = T3(closing_prices, timeperiod=5, vfactor=0)
overlap_df["T3"] = t3
overlap_df.head(10)
```

```python
tema = TEMA(closing_prices, timeperiod=5)
overlap_df["TEMA"] = tema
overlap_df.head(10)
```

```python
trima = TRIMA(closing_prices, timeperiod=5)
overlap_df["TRIMA"] = trima
overlap_df.head(10)
```

```python
wma = WMA(closing_prices, timeperiod=5)
overlap_df["WMA"] = wma
overlap_df.head(10)
```

# Statistic Functions

Reference: https://mrjbq7.github.io/ta-lib/func_groups/statistic_functions.html

```python
stats_df = pd.DataFrame({"Date": df["Date"], "Closing Prices": df["Adj Close"]})
stats_df.head(5)
```

```python
beta = BETA(df["High"], df["Low"], timeperiod=5)
stats_df["BETA"] = beta
stats_df.head(10)
```

```python
correl = CORREL(df["High"], df["Low"], timeperiod=5)
stats_df["CORREL"] = correl
stats_df.head(10)
```

```python
linearreg = LINEARREG(closing_prices, timeperiod=5)
stats_df["LINEARREG"] = linearreg
stats_df.head(10)
```

```python
linearreg_angle = LINEARREG_ANGLE(closing_prices, timeperiod=5)
stats_df["LINEARREG_ANGLE"] = linearreg_angle
stats_df.head(10)
```

```python
linearreg_intercept = LINEARREG_INTERCEPT(closing_prices, timeperiod=5)
stats_df["LINEARREG_INTERCEPT"] = linearreg_intercept
stats_df.head(10)
```

```python
linearreg_slope = LINEARREG_SLOPE(closing_prices, timeperiod=5)
stats_df["LINEARREG_SLOPE"] = linearreg_slope
stats_df.head(10)
```

```python
stddev = STDDEV(closing_prices, timeperiod=5, nbdev=1)
stats_df["STDDEV"] = stddev
stats_df.head(10)
```

```python
tsf = TSF(closing_prices, timeperiod=5)
stats_df["TSF"] = tsf
stats_df.head(10)
```

```python
var = VAR(closing_prices, timeperiod=5, nbdev=1)
stats_df["VAR"] = var
stats_df.head(10)
```
